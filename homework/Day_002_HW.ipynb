```python
import os
import numpy as np
import pandas as pd
```

```python
# Set directory where the data files are saved
dir_data = './data/'
```

```python
# Full file name with path name
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))

# Read csv data into pandas data frame using pd.read_csv
app_train = pd.read_csv(f_app)
```

```python
# Use ? to check the definition of function
?pd.read_csv
```

```python
# 用 .head() 來觀察前5 rows的資料
app_train.head()
```

#### 資料的 row 數以及 column 數

```python
# List data frame row count 
nrow = app_train.shape[0]  # len(app_train)
print("Number of rows in 'application_train.csv' is %s" % (nrow))

# List data frame column count 
ncol = app_train.shape[1]  # len(app_train.columns)
print("Number of columns in 'application_train.csv' is %s" % (ncol))
```

#### 列出所有欄位

```python
print("List all columns of %s " % f_app)
list(app_train)    # app_train.columns
```

#### 截取部分資料

```python
# Show first 5 rows of data frame using .head()
app_train.head()
```

```python
# Show last 5 rows of data frame using .tail()
app_train.tail()
```

```python
# Show 3 random samples of data frame using .sample
app_train.sample(n=3, random_state=1)
```

```python
# Show 1 percent random samples of data frame using .sample
app_train.sample(frac=0.01, random_state=1)
```
